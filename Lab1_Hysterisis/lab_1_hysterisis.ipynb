* R = 1kΩ

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
import scipy.integrate as integrate
import scipy.special as special
```

### Συνάρτηση για διάγραμμα ρεύμα εισόδου Iin(t) - τάση εξόδου Vout(t) ως προς τον χρόνο

```python
def VoutIinonTime(csv,title,Icolor,Vcolor):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv)
    df = df.fillna(method='bfill') # fill the nan values with the next one
    df.rename(columns={'second': 'Time','Volt':'Input Voltage','Volt.1':'Output Voltage'}, inplace=True)
    # Plot the data
    plt.plot(df["Time"], df["Input Voltage"]/1000, label = "Iin(A)",color = Icolor)
    plt.plot(df["Time"], df["Output Voltage"], label = "Vout(t)",color = Vcolor)

    # Add labels to the axes
    plt.xlabel("Time(s)")
    # plt.ylabel("Voltage")
    plt.gcf().set_size_inches(20, 10)

    plt.title(title)
    plt.legend()

    # Show the plot
    plt.show()   
```

### Συνάρτηση για διάγραμμα Βρόχου Υστέρησης (Hysterisis Loop)

```python
def plothysterisis(csv,title,plcolor):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv)
    df = df.fillna(method='bfill')
    df.rename(columns={'second': 'Time','Volt':'Input Voltage','Volt.1':'Output Voltage'}, inplace=True)
    df = df.sort_values(by='Time')

    Vout = (df["Output Voltage"]).to_numpy()
    Iin = (df["Input Voltage"]/1000).to_numpy()
    Time = df["Time"].to_numpy()
    integral = integrate.cumulative_trapezoid(Vout, Time)
    plt.plot(Iin[0:len(integral)], integral,label = "Hysterisis Loop")
    plt.fill_between(Iin[0:len(integral)], integral, 0, alpha=0.2, label='Area',color = plcolor)
    plt.xlabel('Input Current I(A)')
    plt.ylabel('Integral of Output Voltage (dV/dt)')
    plt.title(title)
    plt.legend()
    plt.show()
```

### Συνάρτηση για διάγραμμα Τάσης Εξόδου και Ρεύμα Εισόδου

```python
def VoutIinplot(csv,title,plcolor):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv)
    df = df.fillna(method='bfill')
    df.rename(columns={'second': 'Time','Volt':'Input Voltage','Volt.1':'Output Voltage'}, inplace=True)
    df = df.sort_values(by='Time')

    Vout = (df["Output Voltage"]).to_numpy()
    Iin = (df["Input Voltage"]/1000).to_numpy()
    Time = df["Time"].to_numpy()
    
    fig=plt.figure(figsize=(4,4))
    plt.plot(Iin, Vout,label = "Signal Vin with Iin")
    plt.fill_between(Iin,Vout, label='Area',color = plcolor)
    plt.xlabel('Input Current I (A)')
    plt.ylabel('Output Voltage (V)')
    plt.title(title)
    plt.grid()
```
