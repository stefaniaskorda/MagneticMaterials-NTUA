```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Αισθητήρας Θέσης

### Συνάρτηση για διάγραμμα των αρχείων csv

```python
def plotcsv(csv,title,color):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv)
    df.rename(columns={'second': 'Time(s)','Volt':'Voltage(V)'}, inplace=True)
    # Plot the data
    plt.plot(df["Time(s)"], df["Voltage(V)"], label = "Voltage",color = color)
    # plt.plot(df["Voltage(V)"],df["Time(s)"], label = "Time",color = color)

    # Add labels to the axes
    plt.xlabel("Time(s)")
    plt.ylabel("Voltage(V)")

    plt.gcf().set_size_inches(20, 10)

    plt.title("Σήμα Πηνίου Λήψης - Απόσταση {}".format(title))
    plt.legend()
    plt.axhline(0, color='black', linestyle='-')
    # Show the plot
    plt.show() 
```

### 100mm

```python
plotcsv("100mm.csv","100mm","navy")
```

### 150mm

```python
plotcsv("150mm.csv","150mm","green")
```

### 200mm

```python
plotcsv("200mm.csv","200mm","purple")
```

### 250mm

```python
plotcsv("250mm.csv","250mm","magenta")
```

###  Διάγραμμα απόστασης πηνίων συναρτήσει καθυστέρησης Δt

```python
df_d = [100,150,200,250]
df_dt = [0.214,0.316,0.4285,0.539]
plt.plot(df_d,df_dt,color = "magenta")
# Add labels to the axes
plt.xlabel("Δt(s)")
plt.ylabel("d(mm)")
plt.gcf().set_size_inches(20, 10)
plt.title("Διάγραμμα απόστασης πηνίων συναρτήσει καθυστέρησης Δt")
plt.legend()
plt.axhline(0, color='black', linestyle='-')
# Show the plot
plt.show() 
```
