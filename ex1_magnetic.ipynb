### Μ(Η): μαγνήτιση ως προς το εφαρμοζόμενο πεδίο
#### 1ο Υλικό

```python
m0 = 4*pi*(10**(-7))


def M_field(B,H):
    M = B/m0 - H
    return M
```

```python
b1 = b1.to_numpy()
h1 = h1.to_numpy()
M1 = M_field(b1,h1)
```

```python
figm1 = plt.figure(figsize=(8,3))
axm1 = figm1.add_subplot(1, 1, 1)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
axm1.spines['left'].set_position('center')
axm1.spines['bottom'].set_position('center')

# Eliminate upper and right axes
axm1.spines['right'].set_color('none')
axm1.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
axm1.xaxis.set_ticks_position('bottom')
axm1.yaxis.set_ticks_position('left')
axm1.set_ylabel('M(A/m)')
axm1.set_xlabel('H(A/m)')

axm1.yaxis.set_label_coords(-.1,.5) 
axm1.xaxis.set_label_coords(.5, -.1)

axm1.yaxis.label.set_color('purple')
axm1.xaxis.label.set_color('purple')


plt.plot(h1,M1,color='purple')
plt.title('1ο Υλικό: Μ(Η): μαγνήτιση ως προς το εφαρμοζόμενο πεδίο',color='purple')
# plt.xlabel('H(A/m)',color='magenta')
# plt.ylabel('B(T)',color = 'magenta')
plt.show()
```
